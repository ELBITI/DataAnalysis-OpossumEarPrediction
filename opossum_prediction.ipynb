```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('possum.csv')

#suprimer les lignes qui contient les valeur manquantes
df = df.dropna()

#remplacer les lignes qui contient les valeur manquantes par le moyenne (ca marche si les collonnes contient des variable numérique)
#df.fillna(df.mean(), inplace = True)
print(df.describe())

df.drop_duplicates(inplace=True)
print(df.dtypes)
print(df.describe())

df_selected = df [['footlgth', 'earconch']]


print(df_selected.head())




plt.scatter(df_selected['footlgth'], df_selected['earconch'], color = 'r', marker ='o')


```

```python
#Tache2

def predict_model(alpha, bita, X):

  return alpha*X + bita

def mse(y_reel, y_pred):
  return ((y_reel - y_pred)**2).mean()


X = df_selected['footlgth'].values
Y = df_selected['earconch'].values
plt.scatter(X, Y)
for alpha, beta in [(0.25, 0.5), (0.75, 1), (1,2)]:

  y_pred = predict_model(alpha, beta, X)
  print(f"Model (Alpha={alpha}, beta={beta}) : {mse(y_pred, Y)}")
  plt.plot(X, y_pred, label=f"Model (alpha={alpha}, beta={beta})")

plt.legend()







```
